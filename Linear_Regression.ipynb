#### Linear Regression
#### Linear regression is a statistical and ML technique used to model the relationship between a dependent variable (**label**) and one or more independent variables (**features**) by fitting a straight line to the data. The goal is to predict the dependent variable based on the given independent variables.

### we want to predict a car's fuel efficiency in miles per gallon based on how heavy the car is

```python
import matplotlib.pyplot as plt
import numpy as np

# Dataset
weights = [3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37]  # Car weight in 1000s of pounds
mpg = [18, 15, 18, 16, 15, 14, 24]  # Miles per gallon

# Plotting the data points
plt.scatter(weights, mpg, color='red', marker='o', label="Data Points")

# Labels and title
plt.xlabel("Weight (in 1000s of pounds) (Feature)")
plt.ylabel("Miles per Gallon (MPG) (Gallon)")
plt.title("Car Weight vs Fuel Efficiency")

# Adding the caption
plt.figtext(0.5, -0.1, "Figure 1. Car heaviness (in pounds) versus miles per gallon rating. As a car gets heavier, its miles per gallon rating generally decreases.", wrap=True, horizontalalignment='center', fontsize=10)

# Showing grid and legend
plt.grid(True)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

weights = [3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37]  
mpg = [18, 15, 18, 16, 15, 14, 24] 
plt.scatter(weights, mpg, color='red', marker='o', label="Data Points")

# Computing the best-fit line using numpy's polyfit (degree 1 for linear regression)
slope, intercept = np.polyfit(weights, mpg, 1)

# Creating the best-fit line based on the slope and intercept
best_fit_line = np.array(slope * np.array(weights) + intercept)

# Plotting the best-fit line
plt.plot(weights, best_fit_line, color='blue', label="Best Fit Line")

# Labels and title
plt.xlabel("Weight (in 1000s of pounds) (Feature)")
plt.ylabel("Miles per Gallon (MPG) (Gallon)")
plt.title("Car Weight vs Fuel Efficiency")

# Adjustin axis limits to make the graph more spacious
plt.xlim(min(weights) - 1, max(weights) + 1)  
plt.ylim(min(mpg) - 5, max(mpg) + 5)  

# Adding the caption
plt.figtext(0.5, -0.1, "Figure 2. A best fit line drawn through the data from the previous figure.", wrap=True, horizontalalignment='center', fontsize=10)
plt.grid(True)
plt.legend()
plt.show()
```

### Equation Representation  
\[
y = mx + b
\]  
or in Machine Learning terms:  
\[
y = b + w_1 x_1
\]  
where:  
- \( y \): Predicted output (MPG)  
- \( x_1 \): Feature (Car weight)  
- \( w_1 \): Weight (Slope, learned)  
- \( b \): Bias (Y-intercept, learned)  

### Example Calculation  
For a given model:  
\[
y = 30 + (-3.6) x_1
\]  
Predicting MPG for a **4,000-pound car**:  
\[
y = 30 + (-3.6) \times 4 = 15.6
\]  
Hence, a 4,000-pound car is expected to have 15.6 MPG.  

```python
import matplotlib.pyplot as plt
import numpy as np

weights = np.array([3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37])  
mpg = np.array([18, 15, 18, 16, 15, 14, 24])  

m, b = np.polyfit(weights, mpg, 1)  

x_line = np.linspace(min(weights), max(weights), 100)
y_line = m * x_line + b

x_pred = 4
y_pred = m * x_pred + b  

plt.scatter(weights, mpg, color='red', marker='o', label="Data Points")
plt.plot(x_line, y_line, color='blue', label="Regression Line")

plt.axhline(y=y_pred, color='black', linestyle='--', label=f"y = {y_pred:.2f}")  
plt.axvline(x=x_pred, color='black', linestyle='--', label=f"x = {x_pred}")  
plt.scatter(x_pred, y_pred, color='black', s=100, label=f"Prediction ({x_pred}, {y_pred:.2f})")

plt.xlabel("Pounds in 1000s (Feature)")
plt.ylabel("Miles per Gallon (Label)")
plt.title("Miles per gallon vs. Pounds in 1000s")


plt.figtext(0.5, -0.1, "Figure 3. Using the model, a 4,000-pound car has a predicted fuel efficiency of 15.6 miles per gallon.", wrap=True, horizontalalignment='center', fontsize=10)
plt.grid(True)
plt.legend()
plt.show()
```

#### **Multiple Feature Models in Machine Learning**
##### Model Equation:
\[
y' = b + w_1x_1 + w_2x_2 + w_3x_3 + w_4x_4 + w_5x_5
\]
Where:
- \( y' \): Predicted output (e.g., miles per gallon)
- \( b \): Bias (intercept)
- \( w_i \): Weight for feature \( x_i \)
- \( x_i \): Feature values (e.g., pounds, engine displacement)
for example, 
A model predicting a car’s miles per gallon could include the following features:
- Pounds (car weight)
- Displacement (engine size)
- Acceleration
- Number of cylinders
- Horsepower
### Features:
- \( x_1 \) = Pounds (car weight)
- \( x_2 \) = Displacement (engine size)
- \( x_3 \) = Acceleration
- \( x_4 \) = Number of cylinders
- \( x_5 \) = Horsepower

```python
import matplotlib.pyplot as plt
import numpy as np

# i am using random values for data
displacement = np.random.uniform(70, 450, 100)  
mpg = 35 - 0.05 * displacement + np.random.normal(0, 3, 100)  

plt.figure(figsize=(10, 6))

# Scatter plot
plt.scatter(displacement, mpg, color='red', marker='o', s=30)

# Labels and title
plt.xlabel("Displacement in cubic centimeters\n(Feature)", fontsize=12)
plt.ylabel("Miles\nper gallon\n(Label)", fontsize=12, rotation=0, labelpad=40)


plt.xlim(70, 450)
plt.ylim(10, 40)
plt.figtext(0.5, -0.1, "Figure 4. A car's displacement in cubic centimeters and its miles per gallon rating. As a car's engine gets bigger, its miles per gallon rating generally decreases.", wrap=True, horizontalalignment='center', fontsize=10)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

acceleration = np.random.uniform(8, 20, 100)  
mpg = 10 + 0.5 * acceleration + np.random.normal(0, 3, 100)  
plt.figure(figsize=(10, 4))
plt.gca().set_facecolor('#e0f2f7')

# Scatter plot
plt.scatter(acceleration, mpg, color='red', marker='o', s=30)

# Labels and title
plt.xlabel("Acceleration. 0-60 in seconds\n(Feature)", fontsize=12)
plt.ylabel("Miles\nper gallon\n(Label)", fontsize=12, rotation=0, labelpad=40)

# Adjusting axis limits
plt.xlim(7, 21)
plt.ylim(10, 45)

# Removing gridlines
plt.grid(False)

plt.tight_layout()
plt.show()

# Caption (not part of the plot, but as requested)
print("Figure 5. A car's acceleration and its miles per gallon rating. As a car's acceleration takes longer, the miles per gallon rating generally increases.")
```
